```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("taxis")
```

## Data Analysis

```python
# 6. Calculate average fare by payment type
df.groupby('payment').fare.mean()
```

```python
# 7. Count rides by payment method
df.payment.value_counts()
```

```python
# 8. Calculate total distance by pickup_zone
df.groupby('pickup_zone')['distance'].sum()
```

```python
# 9. Find average tip by dropoff_zone
df.groupby('dropoff_zone')['tip'].mean()
```

```python
# 10. Calculate fare per mile
df['fare_per_mile'] = df['fare'] / df['distance']
df['fare_per_mile']
```

## Time-based Analysis

```python
# 11. Convert pickup column to datetime if needed
df['pickup'] = pd.to_datetime(df['pickup'])
df['pickup']
```

```python
# 12. Extract hour from pickup
df['hour'] = df['pickup'].dt.hour
df['hour']
```

```python
# 13. Average fare by hour
df.groupby('hour')['fare'].mean()
```

```python
# 14. Count rides by day of week
df['day_of_week'] = df['pickup'].dt.day_name()
df['day_of_week'].value_counts()
```

```python
# 15. Monthly ride counts
df['month'] = df['pickup'].dt.month
df.groupby('month').size()
```

## Statistical Analysis

```python

# 17. Calculate percentiles for fare
df['fare'].quantile([0.25, 0.5, 0.75])
```

```python
# 18. Standard deviation of trip distances
df.groupby('pickup_zone')['distance'].std()
```

```python
# 19. Average tip percentage
df['tip_percentage'] = (df['tip'] / df['fare']) * 100
df['tip_percentage']
```

```python
# 20. Identify outliers using IQR for fare
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1
fare_outliers = df[(df['fare'] < (Q1 - 1.5 * IQR)) | (df['fare'] > (Q3 + 1.5 * IQR))]
fare_outliers
```

## Data Filtering

```python
# 21. Rides with tips over $10
high_tips = df[df['tip'] > 10]
high_tips
```

```python
# 22. Long distance rides (over 10 miles)
long_rides = df[df['distance'] > 10]
long_rides
```

```python
# 23. Rush hour rides (8-10 AM)
rush_hour = df[df['pickup'].dt.hour.between(8, 10)]
rush_hour
```

```python
# 24. Weekend rides
weekend_rides = df[df['pickup'].dt.day_name().isin(['Saturday', 'Sunday'])]
weekend_rides
```

```python
# 25. High fare rides (top 10%)
high_fares = df[df['fare'] > df['fare'].quantile(0.9)]
high_fares
```

## Data Aggregation

```python
# 26. Multiple aggregations
df.agg({
    'fare': ['mean', 'min', 'max'],
    'distance': ['mean', 'min', 'max'],
    'tip': ['mean', 'sum']
})
```

```python
# 27. Custom grouping and aggregation
df.groupby('payment').agg({
    'fare': 'mean',
    'tip': 'sum',
    'distance': 'mean'
})
```

```python
# 28. Rolling averages
df.sort_values('pickup')['fare'].rolling(window=10).mean()
```

```python
# 29. Cumulative sums by day
df.groupby('pickup').fare.cumsum()
```

```python
# 30. Pivot table of average fares
pd.pivot_table(df, values='fare', index='pickup_zone', columns='payment', aggfunc='mean')
```

## Advanced Analysis

```python
# 31. Calculate trip duration if dropoff time exists
df['duration'] = (df['dropoff'] - df['pickup']).dt.total_seconds() / 60
df['duration']
```

```python
# 32. Speed calculation (mph)
df['speed'] = df['distance'] / (df['duration'] / 60)
df['speed']
```

```python
# 33. Busy zones analysis
busy_zones = df.groupby('pickup_zone').size().sort_values(ascending=False)
busy_zones
```

```python
# 34. Peak hours identification
peak_hours = df.groupby(df['pickup'].dt.hour).size().sort_values(ascending=False)
peak_hours
```

```python
# 35. Payment method trends over time
payment_trends = df.groupby([df['pickup'].dt.date, 'payment']).size().unstack()
payment_trends
```

## Feature Engineering

```python
# 36. Time period categorization
df['time_period'] = pd.cut(df['pickup'].dt.hour, 
                          bins=[0, 6, 12, 18, 24],
                          labels=['Night', 'Morning', 'Afternoon', 'Evening'])
df['time_period']
```

```python
# 37. Distance categories
df['distance_category'] = pd.cut(df['distance'],
                                bins=[0, 2, 5, 10, float('inf')],
                                labels=['Short', 'Medium', 'Long', 'Very Long'])
df['distance_category']
```

```python
# 38. Fare categories
df['fare_category'] = pd.qcut(df['fare'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
df['fare_category']
```

```python
# 39. Tip percentage categories
df['tip_category'] = pd.cut(df['tip_percentage'],
                           bins=[-float('inf'), 0, 10, 20, float('inf')],
                           labels=['No Tip', 'Low Tip', 'Medium Tip', 'High Tip'])
df['tip_category']
```

```python
# 40. Create rush hour flag
df['is_rush_hour'] = df['pickup'].dt.hour.isin([8, 9, 17, 18, 19])
df['is_rush_hour'] = df['is_rush_hour'].astype(int)
df['is_rush_hour']
```

## Data Cleaning and Preparation

```python
# 41. Remove outliers
df_clean = df[df['fare'].between(df['fare'].quantile(0.01), df['fare'].quantile(0.99))]
df_clean
```

```python
# 42. Handle missing values
df_filled = df.fillna({
    'tip': 0,
    'distance': df['distance'].mean()})
df_filled
```

```python
# 43. Create binary features
df['has_tip'] = df['tip'] > 0
df['has_tip']
```

```python
# 44. Normalize numerical features
df['fare_normalized'] = (df['fare'] - df['fare'].mean()) / df['fare'].std()
df['fare_normalized']
```

```python
# 45. One-hot encode categorical variables
payment_dummies = pd.get_dummies(df['payment'], prefix='payment')
payment_dummies
```

## Summary Statistics and Reports

```python
# 46. Daily summary
daily_summary = df.groupby(df['pickup'].dt.date).agg({
    'fare': ['count', 'sum', 'mean'],
    'tip': ['sum', 'mean'],
    'distance': 'sum'
})
daily_summary
```

```python
# 47. Zone performance metrics
zone_metrics = df.groupby('pickup_zone').agg({
    'fare': ['count', 'mean', 'sum'],
    'tip': ['mean', 'sum'],
    'distance': 'mean'
}).round(2)
zone_metrics
```

```python
# 48. Payment method summary
payment_summary = df.groupby('payment').agg({
    'fare': ['count', 'mean', 'sum'],
    'tip': ['mean', 'sum']
}).round(2)
payment_summary
```

```python
# 49. Time-based metrics
hourly_metrics = df.groupby(df['pickup'].dt.hour).agg({
    'fare': 'mean',
    'tip': 'mean',
    'distance': 'mean'
})
hourly_metrics
```

```python
# 50. Overall performance indicators
performance_metrics = pd.DataFrame({
    'Total_Rides': len(df),
    'Total_Revenue': df['fare'].sum(),
    'Total_Tips': df['tip'].sum(),
    'Avg_Distance': df['distance'].mean(),
    'Avg_Fare': df['fare'].mean(),
    'Tip_Rate': (df['tip'] > 0).mean() * 100
}, index=[0])
performance_metrics
```
