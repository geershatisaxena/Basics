# Visualization of the data

```python
import matplotlib.pyplot as plt
```

Histogram of 'price'

```python
sns.histplot(df['price'], bins=30, kde=True,color = 'g')
plt.title('Distribution of Price')
```

Boxplot of 'price' by 'cut'

```python
sns.boxplot(x='cut', y='price', data=df,color= 'crimson')
plt.title('Price by Cut')
```

Scatterplot of 'carat' vs 'price'

```python
sns.scatterplot(x='carat', y='price', data=df,color='r')
plt.title('Carat vs Price')
plt.show()
```

Violin plot of 'price' by 'color'

```python
sns.violinplot(x='color', y='price', data=df)
plt.title('Price by Color')
```

Countplot of 'cut'

```python
sns.countplot(x='cut', data=df,color = 'mediumvioletred')
plt.title('Count of Diamonds by Cut')
```

```python
df.head()
```

Pairplot of numerical columns

```python
sns.pairplot(df, vars=['carat', 'price', 'depth', 'table'])
```

Barplot of mean 'price' by 'cut'

```python
sns.barplot(x='cut', y='price', data=df, estimator='mean',color = 'lime')
plt.title('Mean Price by Cut')
```

KDE plot of 'carat'

```python
sns.kdeplot(df['carat'], fill=True,color = 'g')
plt.title('Density Plot of Carat')
```

Boxplot of 'price' by 'clarity'

```python
sns.boxplot(x='clarity', y='price', data=df)
plt.title('Price by Clarity')
```

Scatterplot with regression line for 'carat' vs 'price'

```python
sns.regplot(x='carat', y='price', data=df,color = 'crimson')
plt.title('Carat vs Price with Regression Line')
```

Stripplot of 'price' by 'color'

```python
sns.stripplot(x='color', y='price', data=df,color = 'indigo')
plt.title('Price by Color (Stripplot)')
```

Boxenplot of 'price' by 'cut'

```python
sns.boxenplot(x='cut', y='price', data=df,color ='g')
plt.title('Price by Cut (Boxenplot)')
```

Jointplot of 'carat' vs 'price'

```python
sns.jointplot(x='carat', y='price', data=df, kind='scatter')
```

Histogram of 'depth'

```python
sns.histplot(df['depth'], bins=20, kde=True,color = 'blue')
plt.title('Distribution of Depth')
```

FacetGrid of 'price' by 'cut'

```python
g = sns.FacetGrid(df, col='cut', col_wrap=3)
g.map(sns.histplot, 'price', bins=20)
```

Lineplot of mean 'price' by 'carat'

```python
df_grouped = df.groupby('carat')['price'].mean().reset_index()
sns.lineplot(x='carat', y='price', data=df_grouped)
plt.title('Mean Price by Carat')
```

Pie chart of 'cut' distribution

```python
df['cut'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Cut')
plt.ylabel('')
```

Heatmap of pivot table (mean price by 'cut' and 'color')

```python
pivot = df.pivot_table(values='price', index='cut', columns='color', aggfunc='mean')
sns.heatmap(pivot, annot=True, cmap='magma')
plt.title('Mean Price by Cut and Color')
```
