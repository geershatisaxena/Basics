# Visualizing the Data

#### Histogram of orbital periods

```python
sns.histplot(df['orbital_period'], bins=30, kde=True, color='g')
plt.title('Distribution of Orbital Periods')
plt.xlabel('Orbital Period')
plt.ylabel('Frequency')
```

#### Boxplot of planet masses

```python
sns.boxplot(x=df['mass'],color='crimson')
plt.title('Boxplot of Planet Masses')
plt.xlabel('Mass')
```

#### Scatter plot of mass vs orbital period

```python
sns.scatterplot(x='orbital_period', y='mass', data=df,color='goldenrod')
plt.title('Mass vs Orbital Period')
plt.xlabel('Orbital Period')
plt.ylabel('Mass')
```

#### Count plot of discovery methods

```python
sns.countplot(y='method', data=df, order=df['method'].value_counts().index,color='purple')
plt.title('Count of Discovery Methods')
plt.xlabel('Count')
plt.ylabel('Discovery Method')
```

#### Pairplot of numerical columns

```python
sns.pairplot(df,palette='husl', diag_kind='kde', markers='o', hue='method')
plt.suptitle('Pairplot of Numerical Columns', y=1.02)
```

#### Violin plot of mass by discovery method

```python
sns.violinplot(x='method', y='mass', data=df)
plt.xticks(rotation=90)
plt.title('Mass by Discovery Method')
plt.xlabel('Discovery Method')
plt.ylabel('Mass')
```

#### Bar plot of the number of planets discovered per year

```python
sns.barplot(x=df['year'].value_counts().index, y=df['year'].value_counts().values, color='teal')
plt.title('Planets Discovered Per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
```

#### KDE plot of orbital periods

```python
sns.kdeplot(df['orbital_period'], shade=True, color='teal')
plt.title('KDE of Orbital Periods')
plt.xlabel('Orbital Period')
plt.ylabel('Density')
```

#### Line plot of cumulative discoveries over time

```python
df['cumulative_discoveries'] = df['year'].rank(method='dense').astype(int)
sns.lineplot(x='year', y='cumulative_discoveries', data=df,color='teal')
plt.title('Cumulative Discoveries Over Time')
plt.xlabel('Year')
plt.ylabel('Cumulative Discoveries')
```

#### Pie chart of discovery methods

```python
df['method'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(28, 28),colors=sns.color_palette('pastel'))
plt.title('Discovery Methods')
plt.ylabel('')
```

#### Boxplot of orbital period by discovery method

```python
sns.boxplot(x='method', y='orbital_period', data=df,color='crimson')
plt.xticks(rotation=90)
plt.title('Orbital Period by Discovery Method')
plt.xlabel('Discovery Method')
plt.ylabel('Orbital Period')
```

#### Strip plot of mass by discovery method

```python
sns.stripplot(x='method', y='mass', data=df, jitter=True, color='purple')
plt.xticks(rotation=90)
plt.title('Mass by Discovery Method')
plt.xlabel('Discovery Method')
plt.ylabel('Mass')
```

#### Histogram of planet masses

```python
sns.histplot(df['mass'], bins=30, kde=True, color='teal')
plt.title('Distribution of Planet Masses')
plt.xlabel('Mass')
plt.ylabel('Frequency')
```

#### Swarm plot of orbital period by discovery method

```python
sns.swarmplot(x='method', y='orbital_period', data=df,color='purple')
plt.xticks(rotation=90)
plt.title('Orbital Period by Discovery Method')
plt.xlabel('Discovery Method')
plt.ylabel('Orbital Period')
plt.show()
```

#### Line plot of mean mass over years

```python
df.groupby('year')['mass'].mean().plot(kind='line', marker='o', color='r')
plt.title('Mean Mass Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Mass')
```

#### KDE plot of planet masses

```python
sns.kdeplot(df['mass'], shade=True, color='purple')
plt.title('KDE of Planet Masses')
plt.xlabel('Mass')
plt.ylabel('Density')
```

#### Count plot of planets discovered per year

```python
sns.countplot(x='year', data=df, order=df['year'].value_counts().index, color='r')
plt.title('Planets Discovered Per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
```

#### Boxplot of mass by year

```python
sns.boxplot(x='year', y='mass', data=df,color='teal')
plt.xticks(rotation=90)
plt.title('Mass by Year')
plt.xlabel('Year')
plt.ylabel('Mass')
```

#### Scatter plot with regression line for mass vs orbital period

```python
sns.regplot(x='orbital_period', y='mass', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'crimson'},color='teal')
plt.title('Mass vs Orbital Period with Regression Line')
plt.xlabel('Orbital Period')
plt.ylabel('Mass')
```
